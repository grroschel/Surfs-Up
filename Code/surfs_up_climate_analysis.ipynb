# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# engine = create_engine("sqlite:///Resources/hawaii.sqlite")
engine = create_engine("sqlite:///C:/Users/greg_/Surfs-Up/Input/hawaii.sqlite")
```

## This is  where we do the actual reflection

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Another way to look at the table names in this database
from sqlalchemy import inspect  
# Create the inspector and connect it to the engine
inspector = inspect(engine)  
# Collect the names of tables within the database
inspector.get_table_names() 
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## OK; let's get back to work

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
# ----------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------

# Calculate the date 1 year ago from the last data point in the database
# ----------------------------------------------------------------------

# But first, does the 19550 id record exist? Let's look:
cntrec19550 = session.query(Measurement.id,Measurement.station,Measurement.date,Measurement.prcp,Measurement.tobs).filter_by(id=19550).count()  
print(f"The last record, ID 19550, does exist if this is = 1: --> {cntrec19550}")
print()
print("Here's the record ID 19550:")
print("---------------------------")
rec19550 = session.query(Measurement.id,Measurement.station,Measurement.date,Measurement.prcp,Measurement.tobs).filter_by(id=19550).all()  
print(rec19550)
print()
last_date = session.query(Measurement.date).filter_by(id=19550).all()  
print(f"Therefore, the date in the last record is: {last_date[0][0]}")
print()
new_year = (int(last_date[0][0][0:4]) - 1)
new_date = str(new_year) + "-" + str(last_date[0][0][5:7]) + "-" + str(last_date[0][0][8:10])
print(f"Therefore, the date 1 year ago from the last data point in the database is: {new_date}")
print()

# Perform a query to retrieve the data and precipitation scores
# -------------------------------------------------------------

# I assume this means to retrive the date & precipitation amounts
# session.query(Measurement.data,Measurement.prcp).filter_by(date >= new_date).count()  
num_in_new_year = session.query(Measurement.date,Measurement.prcp).filter(Measurement.date >= new_date).count() 
print(f"There are {format(num_in_new_year,',')} observations/records for {new_date} and beyond in our table")
print()
print("Let's look at 20 of them:")
session.query(Measurement.date,Measurement.prcp).filter(Measurement.date >= new_date).limit(20).all()
```

## Save it to a dataframe & sort it

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# ---------------------------------------------------------------------------------

to_pandas = session.query(Measurement.id,Measurement.station,Measurement.date,Measurement.prcp,Measurement.tobs).filter(Measurement.date >= new_date).all()
measurement_df = pd.DataFrame(to_pandas)
measurement_df.head()

# measurement_df_2 = measurement_df.set_index("date")
# measurement_df_2.head()

# Sort the dataframe by date
# --------------------------

measurement_sorted = measurement_df.sort_values("date", ascending=True)
measurement_sorted.head(20)
```

## Dataframe cleanup of NaN Values

```python
print(f"The new dataframe has {format(len(measurement_sorted), ',')} rows")

# Let's check for "NaN" values in the prcp column before we plot
measurement_sorted.count()
measurement_sorted.isnull().any()
```

```python
# Let's get rid of "NaN" values in the prcp column before we plot
measurement_clean = measurement_sorted.dropna(how='any') 
measurement_clean.count()
measurement_clean.isnull().any()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```
